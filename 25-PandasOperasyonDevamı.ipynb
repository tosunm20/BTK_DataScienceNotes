## merge
#### isimler aynı kolonlar farklı (aynı isimlere farklı kolonla veri eklemek)(kaynaştırmak)

```python
mergeSozluk1 = {
    "Isim" : ["Ahmet","Mehmet","Zeynep","Atıl"],
    "Spor" : ["Koşu","Yüzme","Koşu","Basketbol"]
}
```

```python
mergeDataFrame1 = pd.DataFrame(mergeSozluk1)
```

```python
mergeSozluk2 = {
    "Isim" : ["Ahmet","Mehmet","Zeynep","Atıl"],
    "Kalori" : [100,200,150,250]
}
```

```python
mergeDataFrame2 = pd.DataFrame(mergeSozluk2)
```

```python
pd.merge(mergeDataFrame1,mergeDataFrame2, on="Isim")
```

```python
yeniBirVeri = {
    "Karakter Sınıfı" : ["South Park","South Park","Simpson","Simpson"],
    "Karakter Ismi" : ["Cartman","Kenny","Homer","Bart"],
    "Karakter Yas" : [9,10,50,20]
}
```

```python
karakterDF = pd.DataFrame(yeniBirVeri)
```

```python
karakterDF
```

```python
karakterDF.pivot_table(values = "Karakter Yas", index =  ["Karakter Sınıfı", "Karakter Ismi"])   
```

```python
yeniBirVeri2 = {
    "Karakter Sınıfı" : ["South Park","South Park","Simpson","Simpson","Simpson"],
    "Karakter Ismi" : ["Cartman","Kenny","Homer","Bart","Bart"],
    "Karakter Yas" : [9,10,50,20,10]
}
```

```python
karakterDF2 = pd.DataFrame(yeniBirVeri2)
```

```python
karakterDF2.pivot_table(values="Karakter Yas", index = ["Karakter Sınıfı", "Karakter Ismi"])
# Bart için iki deger varken ortalamasını alır
```

```python
karakterDF2.pivot_table(values="Karakter Yas", index = ["Karakter Sınıfı", "Karakter Ismi"],aggfunc = np.sum)
# Bart için iki degeri toplar
```
