```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 2
```

```python
(benimFigur, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1, numpyDizisi2, "g", alpha=1) #alpha ile saydamlık
benimEksen.plot(numpyDizisi2, numpyDizisi1, color="#3A95A8")
```

```python
(yeniFigur,yeniEksen) = plt.subplots()
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 2, color="blue", linewidth = 5.0)
#linewidth = kalınlık değiştirme
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 3, color="yellow", linestyle = ":")

yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 4, color = "#C96F23", linestyle = "-.")

yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 5, color = "#000000", linestyle = "--", marker = "o",markersize=4, markerfacecolor ="r")
# marker değerleri işaretler 1 koyduğunda 2 ise (1,2) noktasını işaretler
```
