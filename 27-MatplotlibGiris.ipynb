```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasListesi = [10,20,30,30,30,40,50,60,70,75]
kiloListesi = [20,60,80,85,86,87,90,95,90,90]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
plt.plot(numpyYasListesi, numpyKiloListesi, "g")
plt.xlabel("Yas Listesi")
plt.ylabel("Kilo Listesi")
plt.title("Kilo'nun yaşa göre değişim grafiği")
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 10
```

```python
plt.plot(numpyDizisi1, numpyDizisi2,"g*-") # g yanına koyulan işaretler tablonun şeklini belirler
plt.xlabel("x")
plt.ylabel("y = x ** 10")
```

```python
plt.subplot(1,2,1) # sıra, grafik sayısı, hangi grafik çizilecek
plt.plot(numpyDizisi1,numpyDizisi2,"r--")
plt.xlabel("x")
plt.ylabel("y=x**10")
plt.subplot(1,2,2)
plt.plot(numpyDizisi2,numpyDizisi1,"r--")
plt.xlabel("x")
plt.ylabel("y=sqrt(x)")
```

```python
benimFigure = plt.figure() # x,y düzlemi eklememiz lazım. düz kagıt gibi düşünülebilir.
figureAxes = benimFigure.add_axes([0.2,0.2,0.5,0.5])
figureAxes.plot(numpyDizisi1,numpyDizisi2)
figureAxes.set_xlabel("X Ekseni Veri İsmi")
figureAxes.set_ylabel("Y Ekseni Veri İsmi")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7]) # sol-sağ, yukarı-aşağı,boyutları ayarlama
eksen2 = figure2.add_axes([0.3,0.3,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2,numpyDizisi1,"r")
eksen2.set_xlabel("X ekseni")
eksen2.set_ylabel("Y ekseni")
eksen2.set_title("2. Grafik Başlık")
```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows=1,ncols=2) # figür, eksen

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2, "g")
    eksen.set_xlabel("X Ekseni")
plt.tight_layout() # görüntü ayarlıyor
```

```python
yeniFigur = plt.figure(figsize=(3,3),dpi=125) # figur boyutu belirleme = figsize=(5,5)
#dpi = dots per inch görüntüyü düzenler
yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi1,numpyDizisi1**2,label= "numpyDizisi^2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1**3, label = "numpyDizisi^3")
yeniEksen.legend(loc=0) # loc = lokasyon belirleme
```

```python
yeniFigur.savefig("benimfigur.png", dpi=200)
```
